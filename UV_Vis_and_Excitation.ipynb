# DFT Calculation to Characterize Molecules

**Density Functional Theory (DFT)** is a quantum mechanical method used to investigate the electronic structure of many-body systems, primarily atoms, molecules, and condensed phases. Instead of solving the many-electron wavefunction directly, DFT simplifies the problem by focusing on the electron density as the fundamental variable. This approach is computationally more efficient and can provide accurate results for a wide range of chemical and physical properties. The core idea of DFT is based on the Hohenberg-Kohn theorems, which assert that the ground-state properties of a system are uniquely determined by its electron density, and that the energy of the system can be minimized through an appropriate functional of the electron density.

**Time-Dependent Density Functional Theory (TDDFT)** is an extension of Density Functional Theory (DFT) that is used to study the excited states of many-body systems. While DFT focuses on the ground state properties, TDDFT is designed to handle time-dependent phenomena and can be used to investigate the response of a system to external perturbations, such as electromagnetic fields.

In TDDFT, the key variable is the time-dependent electron density, which evolves according to the time-dependent Kohn-Sham equations. These equations describe how the electron density changes with time under the influence of an external time-dependent potential. By solving these equations, one can obtain information about the excitation energies, transition probabilities, and dynamic properties of the system.

TDDFT has become a powerful tool in computational chemistry and physics for studying optical properties, electronic excitations, and non-equilibrium processes in molecules, clusters, and solids. Its main advantages are its relatively low computational cost and its ability to provide detailed insights into the electronic dynamics of complex systems.

# Understanding Molecules through UV-Vis and Excitation Spectra

**UV-Vis Spectrum** : The UV-Vis (Ultraviolet-Visible) spectrum is a graphical representation of a molecule's absorption of ultraviolet and visible light. This spectrum provides valuable information about the electronic transitions within a molecule, helping us understand several important aspects of the molecule:

- Electronic Structure: By analyzing the wavelengths at which a molecule absorbs light, we can infer the energy levels of its electrons. Absorption peaks correspond to electronic transitions from lower energy states to higher energy states.

- Functional Groups and Conjugation: Different functional groups and degrees of conjugation in a molecule affect the energy of electronic transitions. For example, π-π* transitions in conjugated systems typically absorb in the UV or visible region, and the position and intensity of these peaks can indicate the extent of conjugation.

- Molecular Environment: The solvent and molecular environment can shift the absorption peaks (bathochromic or hypsochromic shifts) and change their intensities. This provides information about the interactions between the molecule and its surroundings.

- Concentration: According to Beer-Lambert law, the absorbance is directly proportional to the concentration of the absorbing species. Thus, UV-Vis spectroscopy can be used to determine the concentration of a molecule in solution.

**Intensity vs. Excitation Energy** : The intensity vs. excitation energy plot, often referred to as an excitation spectrum, complements the UV-Vis absorption spectrum and provides additional insights:

- Excitation Processes: The excitation spectrum shows the intensity of fluorescence or phosphorescence as a function of the excitation wavelength. This helps identify which electronic transitions are responsible for the observed emission.

- Quantum Yield and Efficiency: The intensity of the emission at different excitation energies can give information about the efficiency of the excitation process and the quantum yield of fluorescence or phosphorescence.

- Energy Transfer: In complex systems, such as molecular aggregates or proteins with multiple chromophores, excitation spectra can help identify energy transfer processes. Peaks in the excitation spectrum corresponding to different chromophores indicate the pathways of energy transfer within the system.

- Photostability: By examining changes in the intensity vs. excitation energy over time or under different conditions, we can assess the photostability of a molecule. Degradation or changes in the spectrum can indicate photobleaching or other photo-induced reactions.

**Understanding Molecules through UV-Vis and Excitation Spectra** : By analyzing UV-Vis absorption spectra and intensity vs. excitation energy plots, scientists can gain a comprehensive understanding of a molecule's:

- Electronic Structure: Identify energy levels and types of electronic transitions.
- Chemical Composition: Determine the presence of specific functional groups and conjugation patterns.
- Molecular Interactions: Study interactions with solvents, other molecules, or the environment.
- Dynamic Processes: Observe energy transfer, excitation efficiency, and photostability.

These insights are crucial for applications in chemistry, biology, materials science, and many other fields, where understanding the electronic and photophysical properties of molecules is essential.

The generated molecule may not be in the most stable configuration (the ground state). Let's implement DFT using PySCF to get the molecule to it's ground state, so that we can easily extract the relevant information about the electronic configuration and it's theoretical orbitals.

```python
os.makedirs(f"Result:DFT_{smiles_representation}", exist_ok=True)

mol = gto.M(
    atom="resources/sample.xyz",
    basis="6-31G",
    symmetry=True,
)
mf = dft.RKS(mol)
mf.xc = "B3PW91"

mf.kernel() * 1.2
e_tot = mf.e_tot * 1.2
mol_eq = optimize(mf)
hessian = mf.Hessian().kernel()
freq_info = thermo.harmonic_analysis(mol, hessian)
vibrational_frequencies = freq_info["freq_wavenumber"]

optimized_xyz_file = "resources/opt_sample.xyz"
mol_eq.tofile(optimized_xyz_file)


def read_xyz_file(xyz_file_path):
    atoms = []
    coords = []
    with open(xyz_file_path, 'r') as file:
        lines = file.readlines()
        for line in lines[2:]:
            parts = line.split()
            atoms.append(parts[0])
            coords.append([float(parts[1]), float(parts[2]), float(parts[3])])
    return atoms, np.array(coords)


def plot_3d_structure(atoms, coords):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    for i, (atom, coord) in enumerate(zip(atoms, coords)):
        ax.scatter(coord[0], coord[1], coord[2], label=atom)
        ax.text(coord[0], coord[1], coord[2], atom, color='red')

    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            distance = np.linalg.norm(coords[i] - coords[j])
            if distance < 1.6:
                ax.plot([coords[i][0], coords[j][0]],
                        [coords[i][1], coords[j][1]],
                        [coords[i][2], coords[j][2]], 'k-')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.savefig(f"Result:DFT_{smiles_representation}/3D_structure.png")
    plt.show()

optimized_xyz_file = "resources/opt_sample.xyz"

atoms, coords = read_xyz_file(optimized_xyz_file)

plot_3d_structure(atoms, coords)

mo_energy = mf.mo_energy * 1.2
n_orbitals = len(mo_energy)
mo_occ = mf.mo_occ * 1.2
n_occ = sum(mo_occ)

homo_index = None
lumo_index = None

for i in range(n_orbitals):
    if mo_occ[i] > 0:
        homo_index = i
    if mo_occ[i] == 0:
        lumo_index = i
        break

if homo_index is not None:
    homo_energy = mo_energy[homo_index]

if lumo_index is not None:
    lumo_energy = mo_energy[lumo_index]

optimized_xyz_file = "resources/opt_sample.xyz"
mol_eq.tofile(optimized_xyz_file)

molden_file = f"Result:DFT_{smiles_representation}/opt_visualization.molden"
with open(molden_file, "w") as f:
    molden.header(mol_eq, f)
    molden.orbital_coeff(mol_eq, f, mf.mo_coeff, ene=mf.mo_energy, occ=mf.mo_occ)

result_file = f"Result:DFT_{smiles_representation}/Results.txt"
with open(result_file, "w") as f:
    f.write(f"Self-Consistent Field Energy: {e_tot}\n")
    f.write("======================================================================\n")
    f.write("Molecular Orbital Energies:\n")
    for i in range(n_orbitals):
        orbital_energy = mo_energy[i]
        f.write(f"Orbital {i + 1}: {orbital_energy * 27.211386} eV\n")
    f.write("======================================================================\n")
    f.write(f"HOMO Energy: {homo_energy * 27.211386} eV")
    f.write(f"\nLUMO Energy: {lumo_energy * 27.211386} eV\n")
    f.write("======================================================================\n")

n_orbitals = len(mo_energy)
orbital_indices = np.arange(1, n_orbitals + 1)
orbital_energies = mo_energy * 27.211386  # Convert to eV

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(orbital_indices, orbital_energies, marker='o', linestyle='-', color='b', label='Orbital Energies')
ax.axhline(y=homo_energy * 27.211386, color='r', linestyle='--', label='HOMO Energy')
ax.axhline(y=lumo_energy * 27.211386, color='g', linestyle='--', label='LUMO Energy')
ax.set_xlabel('Orbital Index')
ax.set_ylabel('Energy (eV)')
ax.set_title('Molecular Orbital Energies')
ax.set_xticks(orbital_indices)
ax.legend()
ax.grid(True)
plt.savefig(f"Result:DFT_{smiles_representation}/orbital energy.png")
plt.show()
```

**Exchange-correlation functionals** are key components in Density Functional Theory (DFT), which describe the combined effects of exchange and correlation interactions among electrons in a system. These interactions are essential for accurately modeling the electronic structure of atoms, molecules, and solids. The exchange-correlation functional is a term in the DFT energy expression that accounts for both:

**Exchange Interaction:** This arises due to the antisymmetry requirement of the electronic wavefunction for fermions (electrons), which leads to a repulsion between electrons with the same spin.
Correlation Interaction: This accounts for the interactions between electrons that are not captured by the mean-field approximation, considering the instantaneous positions and motions of electrons.

**Local Density Approximation (LDA)**

The Local Density Approximation (LDA) is one of the simplest and earliest forms of exchange-correlation functionals used in Density Functional Theory (DFT). In LDA, the exchange-correlation energy at each point in space is assumed to depend only on the electron density at that point. The functional form is derived from the homogeneous electron gas (a system with a uniform electron density), and it is typically expressed as:


lda.jpg

LDA works reasonably well for systems where the electron density varies slowly, such as in many solid-state systems. However, it tends to be less accurate for systems with rapidly varying densities, such as molecules and surfaces.

**Perdew-Burke-Ernzerhof (PBE)**

The Perdew-Burke-Ernzerhof (PBE) functional is a widely used Generalized Gradient Approximation (GGA) exchange-correlation functional. PBE improves upon LDA by including the gradient of the electron density, which allows it to better handle systems with non-uniform densities. The PBE functional is formulated as:

pbe.jpg

PBE is known for providing a good balance between accuracy and computational efficiency and is commonly used in both solid-state and molecular calculations.

**Range-Separated Hybrid Functional: ωB97X-V**

The ωB97X-V functional is a range-separated hybrid functional that combines elements of GGA, exact exchange from Hartree-Fock theory, and empirical dispersion corrections. It includes:

Range-Separation: The exchange interaction is split into short-range and long-range components, treated differently to improve accuracy across different interaction ranges.
Hybrid Exchange: A portion of the exchange energy is derived from exact Hartree-Fock exchange, enhancing the functional's accuracy for systems with significant non-local exchange effects.
Empirical Dispersion Corrections (V): These corrections account for long-range van der Waals interactions, which are important for accurately describing non-covalent interactions and dispersion forces.

ωB97X-V is known for its high accuracy in predicting molecular properties, especially for systems where dispersion forces are significant, such as organic molecules and molecular complexes.

```python
def plot_primitive_gaussian(alpha,label=None):
    r = np.arange(-5,5,0.1)
    g = np.exp(-alpha * r * r)

    plt.plot(r,g, label=label)


def plot_contracted_gaussian(contracted_gaussians, label="contracted"):
    r = np.arange(-5,5,0.1)
    gs = np.zeros_like(r)
    for i in range(1,len(contracted_gaussians[0])):
      gs = np.add(gs,contracted_gaussians[0][i][1] * np.exp(-contracted_gaussians[0][i][0] * r * r))
    gs = gs / np.max(gs)

    plt.plot(r,gs, 'o',label=label)

def plot_slater():
  r = np.arange(-5, 5, 0.1)
  slater_1s = (1/np.sqrt(np.pi))*np.exp(-abs(r))
  slater_1s = slater_1s/np.max(slater_1s)

  plt.plot(r, slater_1s, '*', label='slater')
```

Now, let's plot the stimulated UV-Vis spectrum for our molecule by exciting the molecule in ground state to higher state using TDDFT and extracting useful information.

```python
ha_2_ev = 1 / physical_constants["electron volt-hartree relationship"][0]

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

def run_spectral_analysis(mol, xc="lda"):
    n_states = 15
    spectral_width = 0.1

    mf = dft.RKS(mol, xc=xc).run()

    mytd = tddft.TDDFT(mf)
    mytd.nstates = n_states
    mytd.max_space = 100
    mytd.max_cycle = 200
    mytd.kernel()
    mytd.analyze()
    osc_strengths = mytd.oscillator_strength()[:n_states - 5]

    energies_ev = mytd.e[:n_states - 5] * ha_2_ev
    wavelengths_nm = 1240 / energies_ev

    x_range = np.linspace(wavelengths_nm.min() * 0.9, wavelengths_nm.max() * 1.1, num=1000)
    intensity = np.zeros(x_range.size)

    for e, f in zip(wavelengths_nm, osc_strengths):
        intensity += gaussian(x_range, e, spectral_width) * f

    # Rough Normalization
    dx = (x_range[-1] - x_range[0]) / x_range.size
    area = (intensity * dx).sum()
    intensity /= area

    return x_range, intensity


mol = gto.Mole()
mol.build(
    atom='resources/sample.xyz',
    basis='631g',
    symmetry=True
)

data = {"Wavelength (nm)": []}

xcs = ["lda", "pbe", "wB97X_V"]

for xc in xcs:
    x_range, intensity = run_spectral_analysis(mol, xc=xc)
    data["Wavelength (nm)"] = x_range
    data[f"Intensity ({xc.upper()})"] = intensity

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
for xc in xcs:
    plt.plot(df["Wavelength (nm)"], df[f"Intensity ({xc.upper()})"], label=xc)

plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.title('UV-Vis Spectra for Different Exchange-Correlation Functionals')
plt.legend()
plt.savefig(f"Result:DFT_{smiles_representation}/uv spectrum.jpg")
plt.show()
```

Now, let's plot Intensity Vs Excited Energy, using the information extracted by exciting the molecule to higher state by running TDDFT calculations in the previous cell.

```python
ha_2_ev = 1 / physical_constants["electron volt-hartree relationship"][0]

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

def run_spectral_analysis(mol, xc="lda"):
    n_states = 15
    spectral_width = 0.1

    mf = dft.RKS(mol, xc=xc).run()

    mytd = tddft.TDDFT(mf)
    mytd.nstates = n_states
    mytd.max_space = 100
    mytd.max_cycle = 200
    mytd.kernel()
    mytd.analyze()
    osc_strengths = mytd.oscillator_strength()[:n_states - 5]

    energies_ev = mytd.e[:n_states - 5] * ha_2_ev
    x_range = np.linspace(energies_ev.min() * 0.9, energies_ev.max() * 1.1, num=1000)
    intensity = np.zeros(x_range.size)

    for e, f in zip(energies_ev, osc_strengths):
        intensity += gaussian(x_range, e, spectral_width) * f

    dx = (x_range[-1] - x_range[0]) / x_range.size
    area = (intensity * dx).sum()
    intensity /= area

    return x_range, intensity


mol = gto.Mole()
mol.build(
    atom='resources/sample.xyz',
    basis='631g',
    symmetry=True
)

data = {"Excitation Energy (eV)": [], "Intensity": [], "Exchange-Correlation Functional": []}

xcs = ["lda", "pbe", "wB97X_V"]

for xc in xcs:
    x_range, intensity = run_spectral_analysis(mol, xc=xc)

    data["Excitation Energy (eV)"] += x_range.tolist()
    data["Intensity"] += intensity.tolist()
    data["Exchange-Correlation Functional"] += [xc] * x_range.size

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
for xc in xcs:
    subset_df = df[df["Exchange-Correlation Functional"] == xc]
    plt.plot(subset_df["Excitation Energy (eV)"], subset_df["Intensity"], label=xc)

plt.xlabel('Excitation Energy (eV)')
plt.ylabel('Intensity')
plt.title('Simulated Spectra')
plt.legend()
plt.savefig(
    f"Result:DFT_{smiles_representation}/excitation energy.jpg")
plt.show()
```
